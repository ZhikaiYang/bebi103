### Problem 2.1 (Effects of the prior, 20 pts)

As we have seen and will see in class, it is important to carefully choose the prior distribution in many analysis problems.  However, for many parameter estimation problems where there are many data, the contribution of the prior is overwhelmed by the likelihood, so the exact function form (within some constraints) of the prior is less important.  We will explore this computationally in this problem.

We make many measurements

I will first demonstrate what we are after with an example.  Saw we are trying to determine 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Generate fake data, normally distributed with mu and std.
np.random.seed(42)
real_mean = 1.0
real_std = 1.0
data = np.random.normal(real_mean, real_std, 10)

# Number of data points, for convenience
n = len(data)

# Log posterior for uniform prior
mu = np.linspace(-2.0, 3.0, 200)
sigma = np.linspace(0.01, 3.0, 100)
log_post = np.empty((len(sigma), len(mu)))
for i in range(len(mu)):
    log_post[:, i] = -((data - mu[i])**2).sum() / (2 * sigma**2) \
                        - n * sigma - (2.0 * np.pi) * n / 2.0

# Add a constant to log_post to get it close to unity so well-behaved
# when we exponentiate it.  We don't care about normalization.
log_post -= log_post.max()
```

```python
plt.contourf(mu, sigma, np.exp(log_post))
plt.grid(True)
plt.colorbar()
```

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
mmu, ssigma = np.meshgrid(mu, sigma)
ax.plot_wireframe(mmu, ssigma, np.exp(log_post), rstride=2, cstride=2, lw=0.5)
```
