# Nonlinear regression of Good data

```python
from __future__ import division, absolute_import, \
                                    print_function, unicode_literals

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy.special
import scipy.optimize

# We'll use ColorBrewer for color maps
from brewer2mpl import sequential

# We will also use numdifftools (install using Canopy)
import numdifftools as nd

# Necessary to display plots in this IPython notebook
%matplotlib inline    
```

```python
# Load data into DataFrame
df = pd.read_csv('../data/good_et_al/invitro_droplet_data.csv', 
                 comment='#')

# Rename them all at once
df.columns = ['droplet_diameter', 'droplet_volume', 'spindle_length', 
              'spindle_width', 'tubulin_area']

# Put droplet volume in units of µm^3 for consistency of units
df.droplet_volume *= 1e9
```

```python
# Define the model
def good_non_lin(d, gamma, theta):
    return gamma * d / (1 + (gamma * d / theta)**3)**(1.0 / 3.0)

# Initial guesses
gamma_0 = 0.5
theta_0 = 50.0

# Perform curvefit
popt, junk = scipy.optimize.curve_fit(
    good_non_lin, df.droplet_diameter, df.spindle_length, (gamma_0, theta_0))

# Pull out results
gamma, theta = popt
```

```python
# Compute regression line
d = np.linspace(0.0, 250.0, 200)
spindle_theor = good_non_lin(d, gamma, theta)

# Plot the data points
plt.plot(df.droplet_diameter, df.spindle_length, '.', alpha=0.5, 
         color='#fdbb84')

# Plot the regression line
plt.plot(d, spindle_theor, 'k-')
plt.ylim((15, 50))
plt.xlabel(u'droplet diameter (µm)')
plt.ylabel(u'spindle length (µm)')

# Print parameter values
print('gamma = {0:g}, theta = {1:g}'.format(gamma, theta))
```
