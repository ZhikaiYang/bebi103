I also want my `pretty_activity_plot` function from Tutorial 1a.  If you saved it in a .py file, you can just import the module.  For convenience, I will just redfine it here.

```python
def pretty_activity_plot(ax, x, y, df, xlabel='time (hr)', 
                         ylabel='activity (sec / min)', lw=0.25):
    """
    Makes a pretty plot of sleep traces.  Generates the plot on axes ax,
    and then returns the updated ax.
    """
    # Generate plot (black line if single trace, with colors otherwise)
    if len(y.shape) == 1 or (np.array(y.shape) == 1).any():
        ax.plot(x, y, 'k-', lw=lw)

        # y_max is maximal y value encountered
        y_max = y.max()
    else:
        ax.plot(x, y, '-', lw=lw)

        # Do it this way for compatibility w/ DataFrames and 2d NumPy arrays
        y_max = y.max(axis=0).max()

    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.set_xlim((x.min(), x.max()))
    ax.set_ylim((0.0, y_max))

    # We can overlay day and night.  We'll make night shaded
    ax.fill_between(df.zeit, 0.0, 50*y_max, where=~df.light, 
                    color='lightgray',  alpha=0.5, zorder=0)
    
    return ax
```

## Candidate metrics

### Mean activity

In the Prober, et al. paper, they compute the mean activity for every time point and plot the results for the various genotypes on the same plot.  We can try that.

```python
# Get DataFrames with only data
df_wt_10_act = df_wt_10.drop(['zeit', 'day', 'light'], axis=1)
df_het_10_act = df_het_10.drop(['zeit', 'day', 'light'], axis=1)
df_mut_10_act = df_mut_10.drop(['zeit', 'day', 'light'], axis=1)

# Compute mean activity over time for each genotype
mean_activity_wt = df_wt_10_act.mean(axis=1)
mean_activity_het = df_het_10_act.mean(axis=1)
mean_activity_mut = df_mut_10_act.mean(axis=1)

# Generate y-data for plot
df_plot = pd.concat([mean_activity_wt, mean_activity_het, mean_activity_mut],
                    axis=1)

# Plot the mean activities together
fig, ax = plt.subplots()
ax = pretty_activity_plot(ax, df_wt_10.zeit, df_plot, df_wt_10, lw=1)
plt.legend(('wt', 'het', 'mut'), loc='upper left')
```

As a quick aside, note that this plot is not colorblind safe.  My default color cycle for `matplotlib` works well for me, since I am not colorblind, but if you are preparing plots for publication, you should use a colorblind safe color cycle, if possible.

Here, we seem to see a definite difference between nighttime activity of the homozygous mutant and the wild type and heterozygous mutants.

There are some problems with this analysis, though.  First off, we noted before that there were a couple outliers in the mutant data.  This could artificially tug the nighttime activity of the mutant fish up.  What happens if we re-do the averaging, but we through out the two extreme activity levels at each time point.  (Note: this is not a great way to do outlier detection and correction, as we will talk about later in class.)

```python
# For convenience, get number of samples
n_wt = len(df_wt_10_act.columns)
n_het = len(df_het_10_act.columns)
n_mut = len(df_mut_10_act.columns)

# Adjust means to exclude outliers
mean_activity_adj_wt = (n_wt * mean_activity_wt
        - (df_wt_10_act.max(axis=1) + df_wt_10_act.min(axis=1))) \
        / (n_wt - 2)
mean_activity_adj_het = (n_wt * mean_activity_het
        - (df_het_10_act.max(axis=1) + df_het_10_act.min(axis=1))) \
        / (n_het - 2)
mean_activity_adj_mut = (n_wt * mean_activity_mut
        - (df_mut_10_act.max(axis=1) + df_mut_10_act.min(axis=1))) \
        / (n_mut - 2)

# Generate y-data for plot
df_plot = pd.concat([mean_activity_adj_wt, mean_activity_adj_het, 
                     mean_activity_adj_mut], axis=1)

# Plot the mean activities together
fig, ax = plt.subplots()
ax = pretty_activity_plot(ax, df_wt_10.zeit, df_plot, df_wt_10, lw=1)
lg = plt.legend(('wt', 'het', 'mut'), loc='upper left')
```

The nighttime difference is now not so striking, though, at least for the last two nights, the homozygous *aanat* fish seem to be less restful.  Interestingly, when we delete the extreme measurments, we see a change in the ordering of the activity during waking hours, with the homozygous mutant fish being more active than the heterozygous ones.

### Median activity

We could also plot the median activity over time.  The median is a **robust statistic**, relatively immune to outliers.  So, we compute the median activity for each genotype at each time point.

```python
# Compute medians
median_activity_wt = df_wt_10_act.median(axis=1)
median_activity_het = df_het_10_act.median(axis=1)
median_activity_mut = df_mut_10_act.median(axis=1)

# Generate y-data for plot
df_plot = pd.concat([median_activity_wt, median_activity_het, 
                     median_activity_mut], axis=1)

# Plot the median activities together
fig, ax = plt.subplots()
ax = pretty_activity_plot(ax, df_wt_10.zeit, df_plot, df_wt_10, lw=1)
lg = plt.legend(('wt', 'het', 'mut'), loc='upper left')
```

As in the case where we deleted the extreme activities, we see slightly more active nighttime larvae for the homozygous mutants.  The separation of the daytime trends goes away, suggesting that the distribution of activities at each time point is not symmetric about the mean.

Care has to be taken when plotting the mean or median activity over time.  The larval activity data are very "bursty," meaning that the fish are quiescent for stretches of time, and then dart about.  When averaging over many trajectories, we need to make sure that the window of time over which we sum the active minutes is large enough that we get a decent summary of activity.

It is also important to note that no one fish that will have the mean or median activity.

Presumably the fish have gotten used to their surroundings. We now see that there are subtle differences between the sets (save for the large outliers in the mutant data).  Let's quantify them!

#### Simple Bayesian estimate of the mean

Our measured data are the mean activity over the third night for each fish.  We will not do any extra modeling to get errors for these means, but will instead assume that they all have error coming from a Gaussian distribution with variance $\sigma^2$.  That is to say, we have data that are described by a mean $\mu$ and a variance $\sigma^2$, i.e., each datum is

\begin{align}
x_i = \mu + e_i,
\end{align}

where $e_i$ is the noise component of the $i$the datum.  The noise is assumed to come from a Gaussian distribution with variance $\sigma^2$.  The value of $\sigma$ is the same for all data.  We also assume no prior knowledge about $\mu$ or $\sigma$.

Our goal is to compute $P(\mu~|~D,I)$ for each of the three mutants.  This is the probability that the mean nighttime activity is $\mu$.  The most probable $\mu$ (which is usually the one reported) is where $P(\mu~|~D,I)$ is maximal.

We will derive in lecture on Wednesday (and in section 3.3.1 of Sivia) that in this case, $P(\mu~|~D,I)$ is given by the **Student-t distribution**,

\begin{align}
P(\mu~|~D,I) &\propto \left(1 + \frac{(\mu - \bar{x})^2}{r^2}\right)^{-\frac{n}{2}};\\
r^2 &= \frac{1}{n}\sum_{i\in D}(x_i - \bar{x})^2; \\
\bar{x} &= \frac{1}{n}\sum_{i\in D}x_i,
\end{align}

where $n = |D|$ is the number of data in $D$.

With this formula in hand, we can compute the posterior distributions for the mean mean nighttime activity of each genotype.  In our case, $x_i$ is $N_{A,i}$, the mean nighttime activity for fish $i$.  

(We normally don't care about normalization, but we will normalize the Student-t distribution when we compute it to more easily compare the posteriors for the various genotypes.)

```python
def student_t(mu, x):
    """
    Returns the Student-t distribution for values of mu with data x.
    We could use scipy.stats for this, but we'll do it ourselves.
    """
    # Number of data
    n = len(x)
    
    # Mean of data
    x_mean = x.mean()
    
    # Compute r^2
    r2 = ((x - x_mean)**2).sum() / n
    
    # Compute the mu-dependent part
    t = (1.0 + (mu - x_mean)**2 / r2)**(-n / 2.0)
    
    # Normalize and return
    return scipy.special.gamma(n / 2.0) / np.sqrt(np.pi * r2) \
                / scipy.special.gamma((n-1) / 2.0) * t
```

```python
# Set up values of mu to consider
mu = np.linspace(0.0, 60.0, 300)

# Compute posterior for each of the samples
post_wt = student_t(mu, wt_3rd_night)
post_het = student_t(mu, het_3rd_night)
post_mut = student_t(mu, mut_3rd_night)

# Plot the result
plt.plot(mu, post_wt)
plt.plot(mu, post_het)
plt.plot(mu, post_mut)
plt.xlabel(r'$N_A$ (sec/10 min)')
plt.ylabel(r'$P(N_A|D,I)$')
lg = plt.legend(('wt', 'het', 'mut'), loc='upper right')
```

These posteriors tell a pretty compelling story.  They suggest that the homozygous mutant almost certainly has a higher mean mean nighttime activity than do the wild type and heterozygous mutants.

But we have to be careful that we interpret exactly what the data say.  First, remember that we assumed that the noise in the experiment was Gaussian distributed.  If this is indeed true, then the two very large levels of activity for the homozygous mutant are real measurements; not outliers.  We are then *only* commenting on the *mean* $N_A$ for the larvae.  We know nothing about the distribution of $N_A$ from the posteriors we have plotted here.  Just looking at the data might suggest that some small portion of the homozygous mutants may be prone to being very active at night, but we do not have many data to strongly verify this claim.

If, however, the errors are not strictly Gaussian distributed, it is very possible that the two fish with very large $N_A$ may be outliers.  We will talk about outlier detection and correction later in the class.  For now, just to do a rough investigation, we will see what happens to the posteriors if we delete these two potential outliers.

```python
# Set up values of mu to consider
mu = np.linspace(0.0, 60.0, 300)

# Compute posterior for each of the samples
post_wt = student_t(mu, wt_3rd_night)
post_het = student_t(mu, het_3rd_night)

# Cut out the two largest ones for the mutant
mut_3rd_night.sort(inplace=True)
post_mut = student_t(mu, mut_3rd_night[:-2])

# Plot the result
plt.plot(mu, post_wt)
plt.plot(mu, post_het)
plt.plot(mu, post_mut)
plt.xlabel(r'$N_A$ (sec/10 min)')
plt.ylabel(r'$P(N_A|D,I)$')
lg = plt.legend(('wt', 'het', 'mut'), loc='upper right')
```

We see that even without these potential outliers, the homozygous mutants still have larger $N_A$ than the heterozygous mutants or wild type.

## Bouts

In the Prober, et al. paper, the authors examine "bouts," or extended periods of activity or inactivity.  So, an active bout of length $n$ minutes consists of a fish showing some level of activity during each of $n$ consecutive minutes sampled.  A similar definition applies to rest bouts.  Since the fish move sporaticallys, bout length might be a good metric for wakefulness.

Since we'll need to compute bout lengths for many samples, we'll write a function to do it.

```python
def bout_lengths(s, wake_threshold=0.0):
    """
    Given activity series s, returns two arrays, rest_bouts and active_bouts.
    
    First return value is array of rest bout lengths and second return
    value is array of active bout lengths.
    """
    # Get Boolean for activeness
    active = (s > wake_threshold)
    
    # Check to make sure there is at least one switch
    if np.all(active):
        return np.array([0]), np.array([len(s)])
    elif np.all(~active):
        return np.array([len(s)]), np.array([0])
 
    # Use the NumPy diff function to find indices where switches states
    # switches[i] is the index of first time point in state it switched to
    switches = np.nonzero(np.diff(active))[0] + 1
    
    # Compute bout lengths from switchs, again using np.diff
    bouts = np.concatenate(((switches[0],), np.diff(switches)))
    
    # Find out if active or rest was first and return bout lengths
    if active.iloc[0]:
        return bouts[1::2], bouts[::2]
    else:
        return bouts[::2], bouts[1::2]
```

Now that we can compute bout lengths, let's look at the sleep bout lengths on the third night.  We will use a wake threshold of zero, as in the Prober, et al. paper.  For each fish, we will compute the average length of rest bouts through the night.

```python
# Set up output arrays
wt_mean_bout_length = np.empty(n_wt)
het_mean_bout_length = np.empty(n_het)
mut_mean_bout_length = np.empty(n_mut)

# Get indices for third night
inds = (~df_wt.light) & (df_wt.day==2)

# Compute bout lengths for 3rd night
for i in range(n_wt):
    wt_mean_bout_length[i] = \
            bout_lengths(df_wt[inds][df_wt.columns[i]])[0].mean()
        
for i in range(n_het):
    het_mean_bout_length[i] = \
            bout_lengths(df_het[inds][df_het.columns[i]])[0].mean()
        
for i in range(n_mut):
    mut_mean_bout_length[i] = \
            bout_lengths(df_mut[inds][df_mut.columns[i]])[0].mean()
```

We'll generate jitter plots, as for our previous metric of activity.

```python
# Make beeswarm plot (with jitter in case you couldn't install pybeeswarm)
# Column positions
x_pos = np.arange(3)

# Column labels
x_labels = ['wt', 'het', 'mut']

# Jitter by adding normally distributed random number and plot with 
# alpha < 1, which gives transparency
np.random.seed(42)
x = np.random.normal(x_pos[0], 0.04, size=len(wt_mean_bout_length))
plt.plot(x, wt_mean_bout_length, 'ko', alpha=0.3)
x = np.random.normal(x_pos[1], 0.04, size=len(het_mean_bout_length))
plt.plot(x, het_mean_bout_length, 'ko', alpha=0.3)
x = np.random.normal(x_pos[2], 0.04, size=len(mut_mean_bout_length))
plt.plot(x, mut_mean_bout_length, 'ko', alpha=0.3)
plt.xticks(x_pos, x_labels)
plt.ylabel('mean rest bout length (min)')
plt.ylim((0.0, 5.0))
```

Let's again plot the posteriors describing the true average bout length.

```python
# Set up values of mu to consider
mu = np.linspace(0.0, 4.0, 300)

# Compute posterior for each of the samples
post_wt = student_t(mu, wt_mean_bout_length)
post_het = student_t(mu, het_mean_bout_length)
post_mut = student_t(mu, mut_mean_bout_length)

# Plot the result
plt.plot(mu, post_wt)
plt.plot(mu, post_het)
plt.plot(mu, post_mut)
plt.xlabel(r'mean rest bout length, $\mu$ (min)')
plt.ylabel(r'$P(\mu|D,I)$')
lg = plt.legend(('wt', 'het', 'mut'), loc='upper right')
```

We see that the heterozygous and homozygous mutants have slightly longer rest bouts than the homozygous mutant, but not by much.  We really could not give a firm conclusion on this.
