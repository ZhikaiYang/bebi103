# Nonlinear regression

In this tutorial, we will perform a nonlinear regression on data from a classic paper by Driever and Nüsslein-Volhard.  We will fit the data with an exponential curve plus background.  We will take three approaches.  
1. We will stuff our problem into SciPy's built-in curve fitter, `scipy.optimize.curve_fit`.
2. We will use numerical optimization to find the most likely set of parameters desribing our data under the exponential plus background model.  We will approximate the posterior probability as a two-dimensional Gaussian and use numerical differentiation to compute the covariance matrix.  This is all done under the hood using JB's `curve_fit_1d` module.
3. For our third regression, we will use the `emcee` package to sample the posterior distribution using Markov chain Monte Carlo.

We assume our data have the form of an exponential plus background,

\begin{align}
y(x) = a + b\mathrm{e}^{-x/\lambda},
\end{align}

where $y(x)$ is the *bcd* intensity from immunostaining measurements and $x$ is the position along the AP axis of the fly embryo (in units of the total AP axis length).  This is our *hypothesis*, $H$, that the data, sans error, follow this functional form.  We have three regression parameters under this hypothesis, $a$ $b$, and $\lambda$, which we will collectively call $\mathbf{a} = \{a, b, \lambda\}$.

This tutorial is inspired by [this great post](http://jakevdp.github.io/blog/2014/06/14/frequentism-and-bayesianism-4-bayesian-in-python/) on the blog [Pythonic Preambulations](http://jakevdp.github.io) by Jake Vanderplas.

## Naïvely using `scipy.optimize.curve_fit`

The `scipy.optimize` module has a function called `curve_fit` designed to find parameters that best fit $(x,y)$ data to a prescribed function.  The function definition is:
```
opt.curve_fit(f, xdata, ydata, p0=None, sigma=None, \
              absolute_sigma=False, **kw)
```

The doc string tells us what is going on with this:
```
Assumes ``ydata = f(xdata, *params) + eps``
...
f : callable
    The model function, f(x, ...).  It must take the independent
    variable as the first argument and the parameters to fit as
    separate remaining arguments.
```

The input `p0` is an array containing our best guess as to what the parameters are.  Let's use `scipy.optimize.curve_fit` to perform our regression.

### What `scipy.optimize.curve_fit` does

`scipy.optimize.curve_fit` uses the Levenberg-Marquart algorithm for solving least-squares problems as implemented in the `scipy.optimize` module.  It minimizes the sum of the squares of the residuals.  I.e., it solves

\begin{align}
\arg \min_{\mathbf{a}} \sum_{i\in D} \frac{(y_i - y(x_i,\mathbf{a}))^2}{\sigma_i^2},
\end{align}

where $D = \{(x_0, y_0), (x_1, y_1), \ldots\}$ is the set of data points we have measured and $\sigma_i$ is the error in measurement of data point $i$.  $y(x_i,\mathbf{a})$ is the value of $y$ predicted by the model at point $x_i$ with parameter set $\mathbf{a}$.  The right hand side of the above equation is often referred to as the $\chi^2$ statistic. 

**Importantly, if the $\sigma_i$'s are not known, `scipy.optimize.curve_fit` assumes $\sigma_i = 1\;\forall i \in D$.**

Note that the L-M algorithm only will find local minima in $\chi^2$.  Therefore, it is important to get as good of an initial guess of parameters as possible.

After performing the curve fit, it assumes that the errors in the fit parameters are approximately Gaussian and returns a covariance matrix.

### Defining our model function

We have to make sure the function is set up precisely as defined in the doc string.

````python
def y_model_scipy(x, a, b, lam):
    return a + b * np.exp(-x / lam)
````

### Making the initial guess

The initial guess must be specified as a `numpy` array.  We would like to give a good guess.  We know that $y(x\to\infty) = a$, we we guess $a$ to be the minimum value of our data.  We also have $y(0) = a + b$, so we guess $b$ to be the largest value of the data minus our guess for $a$.  Finally, $\lambda$ is the length over which the signal drops about $2/3$ of the way toward its asymptotic value.  So, we guess $\lambda \approx 0.2$.

````python
# We order parameters: a, b, lambda.  This is the order they'll be returned.
p0 = np.array([data.y.min(), data.y.max() - data.y.min(), 0.2])
````

### Running `scipy.optimize.curve_fit`

The function returns two `numpy` arrays.  The first is the optimal parameter values and the second is the approximate covariance matrix.

````python
# Run curve_fit
popt, pcov = curve_fit(y_model_scipy, data.x, data.y, p0=p0)

# Parse the output
a_star_sp, b_star_sp, lam_star_sp = popt
a_err_bar_sp, b_err_bar_sp, lam_err_bar_sp = np.sqrt(np.diag(pcov))

# Print the results
print(
"""
a      = {0:.3f} +- {1:.3f}
c      = {2:.3f} +- {3:.3f}
lambda = {4:.3f} +- {5:.3f}
""".format(a_star_sp, a_err_bar_sp, b_star_sp, b_err_bar_sp,
           lam_star_sp, lam_err_bar_sp))
````

We can plot the best fit line with the optimal parameters.

````python
# Generate a densely populated array of x value for smooth plots
x_smooth = np.linspace(data.x.min(), data.x.max(), 200)

y_smooth = y_model_scipy(x_smooth, *popt)

plt.plot(data.x, data.y, 'ko', x_smooth, y_smooth, 'r-')
plt.ylim((0.0, 1.0))
plt.xlabel(r'$x$')
plt.ylabel(r'$y$');
````

## Constructing the Bayesian problem

We just stuffed our model function into `scipy.optimize.curve_fit` and let it go to work.  For the other two regression methods we will employ, we will more carefully construct the problem in a Bayesian framework.

### Defining the likelihood

Our goal in regression is to compute relevant properties of the posterior probability, $P(\mathbf{a}~|~D,H)$, where again $D = \{(x_0, y_0), (x_1, y_1), \ldots\}$ is the set of data points we have measured and $H$ is our hypothesis, in this case that the data follow a exponential curve plus background.  In particular, we would like to know $\textbf{a}^*$, the most likely set of parameters.  We would also like to know how broad the distribution is.  That is, we would like to know how probable $\textbf{a}^*$ is compared to other values of $\mathbf{a}$.

We start by defining the likelihood.  We assume the error in the data are Gaussian with standard deviation $\sigma$.  We assume that each measurement is independent of the others and that every measurement is subject to the same $\sigma$.  These are the assumptions that are under the hood in JB's `curve_fit_1d` module, so we do not have to specify them specifically there, but we will when using `emcee` for MCMC.  Under these assumptions, the likelihood is

\begin{align}
P(D~|~\mathbf{a},\sigma,H) = \prod_{i\in D} \frac{1}{\sqrt{2\pi\sigma^2}}\,\exp\left\{-\frac{(y_i - y(x_i,\mathbf{a}))^2}{2\sigma^2}\right\},
\end{align}

where $y(x_i, \mathbf{a})$ is the value of $y$ at position $x_i$ as predicted by the biexponential curve.  We need to specify the log likelihood to do MCMC, so we compute this as

\begin{align}
L \equiv \ln P(D~|~\mathbf{a},\sigma,H) = \text{constant} -\frac{|D|}{2}\,\ln \sigma^2 - \frac{1}{2\sigma^2} \, \sum_{i\in D} (y_i - y(x_i,\mathbf{a}))^2.
\end{align}

### Defining the prior

Next, we define the prior.  If we have many data points, the precise definition of the prior is not as important, as the likelihood dominates the prior.  Nonetheless, we want to do the best we can for defining the prior.  

We want to assume we know next to nothing about the parameters.  I saw next to nothing because we do know that $a > 0$.  We know this by construction of our model; i.e., $a$ is a background signal and Driever and Nüsslein-Volhard did not do any background subtraction, so it must be positive.  We also know that under our hypothesis, $y(0) = a + b$.  Physically, $y > 0$, so we know $b > -a$.  This constraint is more difficult to treat, so we define a new parameter $c = a + b$ which must be positive.  Our new model is then

\begin{align}
y(x) = a + (c - a)\mathrm{e}^{-x/\lambda}.
\end{align}

So, only considering known physical constraints, we have $a,c > 0$.

If we look at the data, we see that the decay length $\lambda$ is obviously positive.  However, *without* looking at the data, we could not say this.  If the embryo were infinite, we could say $\lambda$ must be positive because we know that physically the measured immunostaining intensity must be finite.  But this is not the case, so $\lambda$ may be either positive or negative.  Remember, we cannot use information from our data in defining the prior, $P(\mathbf{a},\sigma~|~H)$.

We also have our nuisance parameter, $\sigma$, and we know that $\sigma > 0$.  Other than that, we want to claim maximal ignorance about our paramters.  So, we choose uniform priors for $\lambda$, and Jeffreys priors for $a$, $c$, and $\sigma$.

\begin{align}
P(\mathbf{a},\sigma~|~H) \propto \frac{1}{a c \sigma},
\end{align}
giving a log prior of
\begin{align}
\log P(\mathbf{a},\sigma~|~H) = \text{constant} - \ln a - \ln c - \ln \sigma.
\end{align}

Note that uniform priors are not necessarily the least informative.  The choice of prior is dependent on the model function we use, and that choice is seldom trivial.  For now, we leave it at two possible choices, either a uniform or Jeffreys prior.

## Regression using `curve_fit_1d`

Now that we have our function defined, we can use JB's curve fitting tool to compute the most likely set of parameters.

### What `curve_fit_1d` does

`curve_fit_1d` takes a Bayesian approach to finding the most probable set of parameters, including the appropriate priors and integrating out the nuisance parameter $\sigma$.  Finding the most probable parameters is done without approximation.  However, once the most probable parameters are found, the posterior distribution is approximated as Gaussian.  It will definitely be Gaussian at the maximum, as this is the result from performing a Taylor series about the maximal point.  But it could deviate from Gaussian behavior away (and possibly not too far away) from the maximum.

`curve_fit_1d` uses the `scipy.optimize` module to find the parameters that maximize the posterior probability.  The precise algorithm depends on some input parameters, but it generally uses techniques of convex optimization, which means that the guess of the parameters is important.  `curve_fit_1d` will only find a local maximum in the posterior.  Upon finding the optimal parameters, `curve_fit_1d` numerically differentiates the posterior (effectively determining the coefficients of a Taylor series approximation) to compute the Gaussian exproximation of the posterior.

### Location and scale parameters and defining the model

One trick to consider is that `curve_fit_1d` assumes all variables are location parameters, meaning that we should use the logarithm of any parameter that has a Jeffreys prior.  We therefore need to define our exponential plus background function as a function of the logarithms of $a$ and $c$.

````python
def y_model_cf1d(p, x):
    """
    Returns an exponential function plus background based on parameters in p.
    """
    log_a, log_c, lam = p
    a = np.exp(log_a)
    c = np.exp(log_c)
    return a + (c - a) * np.exp(-x / lam)
````

### The initial guess

Now that our function is defined, we have to come up with initial guesses for $a$, $c$ and $\lambda$ to feed into `curve_fit_1d`.  With our new function definition, $y(0) = c$, so we guess $c$ to be the value of the first data point.  We also have $y(x\to\infty) = a$, so we guess $a$ to be the value of our last data point.  Finally, as before, we guess $\lambda \approx 0.2$.

````python
# Initial guess; remember: our parameters are log(a), log(c), and lambda
p0 = np.array([np.log(data.y.max()), np.log(data.y.min()), 0.2])
````

These are good guesses because out curve has a simple shape.  It might be harder to generate good intital guess.  Remember that `curve_fit_1d` needs good guesses!

### Running `curve_fit_1d`

We can now run `curve_fit_1d` to compute the optimal values of the parameters and their.  To compute the error bars, `curve_fit_1d` approximates the posterior probablity as locally Gaussian at $\mathbf{a}^*$.  The covariance is calculated by numerically computing second derivatives at $\mathbf{a}^*$.  The error bar people typically report is given by the square root of the self-variances of each parameter (corresponding to an error bar of $\pm$ one standard deviation).  For scale parameters, we report the error as `*/` since we computed the covariance of the log of the parameter.

````python
# Run the curve fit; return results to popt, pcov
popt, pcov = cf.curve_fit_1d(y_model_cf1d, data.x, data.y, p0, quiet=True)

# Pull the results out for convenience
a_star, c_star = np.exp(popt[:2])
lam_star = popt[2]
a_err_bar, c_err_bar = np.exp(np.sqrt(np.diag(pcov)[:2]))
lam_err_bar = np.sqrt(pcov[2,2])

# Print the result
print(
"""
a      = {0:.3f} */ {1:.3f}
c      = {2:.3f} */ {3:.3f}
lambda = {4:.3f} +- {5:.3f}
""".format(a_star, a_err_bar, c_star, c_err_bar, lam_star, lam_err_bar))
````

We see that all of the parameters have pretty tight error bars.  We can now plot the curve with the most probable parameters with the data.

````python
# Generate a densely populated array of x value for smooth plots
x_smooth = np.linspace(data.x.min(), data.x.max(), 200)

y_smooth = y_model_cf1d(popt, x_smooth)

plt.plot(data.x, data.y, 'ko', x_smooth, y_smooth, 'r-')
plt.xlabel(r'$x$')
plt.ylabel(r'$y$');
````

### Comparison of results with `scipy.optimize.curve_fit`

These results are very similar to those obtained by using `scipy.optimize.curve_fit`.  We can compare them side-by-side.  To make the comparison, I convert the `*/` error bars to approximate `+-`.  We also convert between the parameter $b$ to $c$.

````python
# Convert parameter b from scipy.optimize.curve_fit to c with 
# error propagation
c_star_sp = b_star_sp + a_star_sp
c_err_bar_sp = np.sqrt(a_err_bar_sp**2 + b_err_bar_sp**2)

print("""
          curve_fit         JB's curve_fit_1d
        ==============       ==============
a      = {0:.3f} +- {1:.3f},     {6:.3f} +- {7:.3f}
c      = {2:.3f} +- {3:.3f},     {8:.3f} +- {9:.3f}
lambda = {4:.3f} +- {5:.3f},     {10:.3f} +- {11:.3f}
""".format(a_star_sp, a_err_bar_sp, c_star_sp, c_err_bar_sp, 
           lam_star_sp, lam_err_bar_sp,
           a_star, np.log(a_err_bar), c_star, np.log(c_err_bar), 
           lam_star, lam_err_bar))
````

We see that we get similar result between the two, with the exception of a larger error bar on the parameter $a$ using `curve_fit_1d`.  Because `scipy.optimize.curve_fit` makes the erroneous assumption that all $\sigma_i$ in the expression for the $\chi^2$ statistic are unity, it will report erroneous error bars for some parameters, particularly when the data are of lower quality.  A more appropriate strategy is to marginalize the error in data points as a nuisance parameter.

Note that the optimal parameter values are provably identical between the two methods.  They only differn in the error bars.  However, as we will see in the next analysis, for multivariate regression, the error bars only tell part of the story.

### Analysis of the approximate posterior probability

It would also be useful to plot the posterior distribution.  In this optimization treatment, we approximate the posterior as Gaussian.  Specifically,

\begin{align}
P(\mathbf{a}~|~D,H) \approx \frac{1}{\sqrt{(2\pi)^3 |\mathsf{\Sigma}|}} \, \exp\left\{-(\mathbf{a} - \mathbf{a}^*)^\mathsf{T} \cdot \mathsf{\Sigma}^{-1} \cdot (\mathbf{a} - \mathbf{a}^*)^\mathsf{T}\right\},
\end{align}

where $\mathsf{\Sigma}$ is the covariance matrix.  With this approximation, we can compute, for example,

\begin{align}
P(a, \lambda~|~D,H) &\approx \frac{1}{\sqrt{(2\pi)^3 |\mathsf{\Sigma}|}} \int_{-\infty}^\infty \mathrm{d}c\, \exp\left\{-\frac{1}{2}(\mathbf{a} - \mathbf{a}^*)^\mathsf{T} \cdot \mathsf{\Sigma}^{-1} \cdot (\mathbf{a} - \mathbf{a}^*)\right\} \\
&= \frac{1}{\sqrt{(2\pi)^2 |\mathsf{\Sigma}_{a\lambda}|}} \, \exp\left\{-\frac{1}{2}(\mathbf{a}_{a\lambda} - \mathbf{a}_{a\lambda}^*)^\mathsf{T} \cdot \mathsf{\Sigma}_{a\lambda}^{-1} \cdot (\mathbf{a}_{a\lambda} - \mathbf{a}_{a\lambda}^*)\right\}.
\end{align}

In other words, to get the two-dimensional distribution of $a$ and $\lambda$, we marginalize $c$.  Conveniently, the result of marginalization of a multidimensional Gaussian is itself a Gaussian.  We denote with the subscript $c\lambda$ the portion of the respective matrices or vectors pertaining to $c$ and $\lambda$.  We can then compute the approximate posterior.

````python
# Indices of a and lambda
inds = np.array([0,2])

# Most probable values of a and lambda
log_a_star, lam_star = popt[inds]

# Slice out the pertinent part of the covariance matrix
cov = pcov[inds][:,inds]

# Inverse and determinant of the pertinent part of the covariance matrix
icov = la.inv(cov)
dcov = la.det(cov)

# Ranges of parameters to plot
log_a_smooth = np.linspace(np.log(0.15), np.log(0.21), 100)
lam_smooth = np.linspace(0.15, 0.22, 100)

# Make a meshgrid version to generate the posterior
log_aa, ll = np.meshgrid(log_a_smooth, lam_smooth)

# Argument of exponential, easiest if we just write out the quadratic form
quad_form = -icov[0,0] * (log_aa - log_a_star)**2 \
            - (icov[1,0] + icov[0,1]) \
                  * (log_aa - log_a_star) * (ll - lam_star) \
            - icov[1,1] * (ll - lam_star)**2

# Generate posterior
approx_posterior = np.exp(quad_form / 2.0) / np.sqrt((2.0 * np.pi)**2 * dcov)

# Use one of the ColorBrewer2 maps
bmap = brewer2mpl.get_map('YlGnBu', 'sequential', 9, reverse=True)
cmap = bmap.get_mpl_colormap(N=9, gamma=1.0)

# Plot the contours
plt.contourf(np.exp(log_a_smooth), lam_smooth, approx_posterior, cmap=cmap)
plt.xlabel(r'$a$');
plt.ylabel(r'$\lambda$');
````

While this plot of the posterior is useful, it is perhaps more useful to construct credible regions.  Since the posterior is Gaussian, the credible regions correspond to the familiar confidence region.  We wish to find the contour within which a fraction $\alpha$ of the total probability lies.  We could numerically integrate the posterior, and this is in essence what we do when we do MCMC, but we can take advantage of the Gaussian nature of our posterior.  We recall that the posterior probability distribution is

\begin{align}
P(a, \lambda~|~D,H) &= \frac{1}{\sqrt{(2\pi)^2 |\mathsf{\Sigma}|}} \, \exp\left\{-\frac{1}{2}(\mathbf{a} - \mathbf{a}^*)^\mathsf{T} \cdot \mathsf{\Sigma}^{-1} \cdot (\mathbf{a} - \mathbf{a}^*)\right\},
\end{align}

where we have dropped the $a\lambda$ subscripts with the understanding that we are considering the bivariate distribution.  Because the covariance matrix is symmetric positive define, we can write

\begin{align}
\mathsf{\Sigma} = \mathsf{Q} \cdot \mathsf{\Lambda} \cdot \mathsf{Q}^\mathsf{T},
\end{align}

where $\mathsf{Q}$ is an orthogonal matrix comprized of the normalized eigenvectors of $\mathsf{\Sigma}$ and

\begin{align}
\mathsf{\Lambda} = \begin{pmatrix}\lambda_1 & 0 \\
0 & \lambda_2\end{pmatrix}
\end{align}

is the diagonal matrix of the eigenvalues of $\mathsf{\Sigma}$.  Taking advantage of the orthogonality of $\mathsf{Q}$ ($Q^\mathsf{T} = Q^{-1}$), we can rewrite the quadratic form in the the argument of the exponential as

\begin{align}
(\mathbf{a} - \mathbf{a}^*)^\mathsf{T} \cdot \mathsf{\Sigma}^{-1} \cdot (\mathbf{a} - \mathbf{a}^*)
&= (\mathbf{a} - \mathbf{a}^*)^\mathsf{T} \cdot (\mathsf{Q} \cdot \mathsf{\Lambda} \cdot \mathsf{Q}^\mathsf{T})^{-1} \cdot (\mathbf{a} - \mathbf{a}^*) \\
&=(\mathbf{a} - \mathbf{a}^*)^\mathsf{T} \cdot \mathsf{Q} \cdot \mathsf{\Lambda}^{-1} \cdot \mathsf{Q}^\mathsf{T} \cdot (\mathbf{a} - \mathbf{a}^*) \\
&= (Q^\mathsf{T}\cdot(\mathbf{a} - \mathbf{a}^*))^\mathsf{T} \cdot \mathsf{\Lambda}^{-1} \cdot \mathsf{Q}^\mathsf{T} \cdot (\mathbf{a} - \mathbf{a}^*) \\
&= \mathbf{u}^\mathsf{T} \cdot \mathsf{\Lambda}^{-1} \cdot \mathbf{u},
\end{align}

where we have defined a new variable $\mathbf{y} = \mathsf{Q}^\mathsf{T} \cdot (\mathbf{a} - \mathbf{a}^*)$.  Using the fact that $|\mathsf{\Sigma}| = \lambda_1 \lambda_2$, we can rewrite the posterior as

\begin{align}
P(y_1, y_2~|~D,H) = \left[\frac{1}{\sqrt{2\pi \lambda_1}}\,\mathrm{e}^{-y_1^2~/~2\lambda_1}\right]\,
\left[\frac{1}{\sqrt{2\pi \lambda_2}}\,\mathrm{e}^{-y_2^2~/~2\lambda_2}\right].
\end{align}

We have now almost magically completely decoupled the Gaussian distributions of the two random variables $a$ and $\lambda$.  When plotted in the $(y_1/\sqrt{\lambda_1}, y_2/\sqrt{\lambda_2})$, the posterior is completely symmetric.  We can already exploit what we know about Gaussians.  We will get $\alpha \approx 0.68$ of the total probability if $y_1 < \sqrt{\lambda_1}$ and $y_2 < \sqrt{\lambda_2}$ and $\alpha \approx 95$ of the total probability if $y_1 < 2\sqrt{\lambda_1}$ and $y_2 < 2\sqrt{\lambda_2}$.  We can compute and plot the contours.

````python
# Define a func for the orthogonalized posterior (function of y and lambda)
def orthogonal_posterior(y, lam):
    return np.prod(np.exp(-y**2 / 2.0 / lam)) \
                     / (np.sqrt(2.0 * np.pi)**len(y) * np.sqrt(np.prod(lam)))

# Compute the eigenvalues of the covariance matrix
lam = la.eigvals(cov)

# Find the level of the posterior and 1 and 2 stdevs
post_1std = orthogonal_posterior(np.sqrt(lam), lam)
post_2std = orthogonal_posterior(2.0 * np.sqrt(lam), lam)

# Overlay the contours
# I got the color from messing with "road" colors on ColorBrewer2.
levels = [post_1std, post_2std]
plt.contourf(np.exp(log_a_smooth), lam_smooth, approx_posterior, cmap=cmap)
contours = plt.contour(np.exp(log_a_smooth), lam_smooth, approx_posterior, 
                       colors='#ff00d7', levels=levels)

# Label the levels
fmt = {}
fmt[levels[0]] = '68%'
fmt[levels[1]] = '95%'
plt.clabel(contours, fmt=fmt)

# For the sake of illustration we will show our reported +/- conf intervals
a_bar = [np.exp(log_a_star - np.sqrt(cov[0,0])),
         np.exp(log_a_star + np.sqrt(cov[0,0]))]
lam_bar = [lam_star - np.sqrt(cov[1,1]),
           lam_star + np.sqrt(cov[1,1])]
plt.plot(a_bar, [lam_star, lam_star], color='#ff00d7', solid_capstyle="butt")
plt.plot(np.exp(np.array([log_a_star, log_a_star])), lam_bar, 'k-',
         color='#ff00d7', solid_capstyle='butt')

plt.xlabel(r'$a$')
plt.ylabel(r'$\lambda$');
````

### Conclusions from using `curve_fit_1d`

`curve_fit_1d` gave us the most probable set of parameters describing the data, given that our hypthesis $H$ (that the data are following an exponential decay plus background) is true.  The meaning of the error bars is less clear.  The error bar is typically interpreted to mean that the parameter has a 68% chance of being within the bounds of the error bar.  As we can see from the above plot, in a multi-dimensional regression, this meaning could be lost.  If, for example, there were big uncertainty in the value of $a$, we would in turn be less certain of $\lambda$, though the orthogonal error bar would not fully capture that.  This is because the values of $a$ and $\lambda$ are dependent on each other.

## Regression using MCMC

An alternative approach to regression is to compute the posterior probability by repeatedly sampling it.  We won't get into the theory here (that's for lecture!).  Instead, we will use the package `emcee` (a.k.a. MCMC Hammer) to perform the MCMC sampling.  

`emcee` is really easy to use.  We just have to provide a function defining the log of the posterior.

That said, I think for `pymc` is the preferred.  It is more complete and more powerful package.  However, it will soon be having a major upgrade to `pymc3`, which will have new APIs.  Since `emcee` is so easy to use and we don't want to waste time learning soon-to-be-deprecated APIs, we'll stick with `emcee`.


### Defining the log of the posterior

And so we being by defining the posterior.  I find it's most intuitive to split it into the likelihood and the prior.

````python
def log_likelihood(p_with_sigma, f, x, y, f_args):
    """
    Returns the log of the posterior consisting of the product of Gaussians.
    
    The last entry in p_with_sigma is the nuisance parameter describing 
    error in measurement.
    """
    n = len(x)
    y_model = f(p_with_sigma[:-1], x, *f_args)
    sigma = p_with_sigma[-1]
    return -n / 2.0 * np.log(sigma**2) \
            - np.sum((y - y_model)**2) / 2.0 / sigma**2

def log_prior(p, jeffreys):
    """
    Returns log of prior where p[i] is Jeffreys prior if jeffreys[i] is True.
    Otherwise, p[i] is assumed to be a uniform prior.
    """
    if np.any(p[jeffreys] < 0.0):
        return -np.inf
    else:
        return -np.sum(p[jeffreys])
    
def log_posterior(p_with_sigma, f, x, y, f_args, jeffreys):
    """
    Returns the log of the posterior (sans an additive constant)
    """
    return log_prior(p_with_sigma, jeffreys) \
            + log_likelihood(p_with_sigma, f, x, y, f_args)
````

I wrote these functions so that they are general.  They will work for any problem where the data points are assumed to be independent and do not have error bars, but the errors are assumed to be Gaussian.  We just need to specify the function `f` that describes the hypothesis we are using in the regression.

### Defining the model

Note that unlike when we used `curve_fit_1d`, we explicitly take into account parameters that have Jeffreys priors; we do not convert everything to location parameters by taking logarithms.  That means we need to specify a new model function, different from the one we used with `curve_fit_1d`, for `f`.

````python
def y_model(p, x):
    """
    Returns an exponential function plus background based on parameters in p.
    """
    a, c, lam = p
    return a + (c - a) * np.exp(-x / lam)
````

### Setting up the `emcee` calculation

We now have all the pieces of our data and model set up for MCMC calculation.  We need to decide how many walkers we want and how many steps they will take.  This is kind of an arbitrary choice.  We will choose 50 walkers with 1000 burn-in steps followed by 2000 MCMC steps.

````python
n_dim = 4  # number of parameters in the model
jeffreys = np.array((True, True, False, True)) # order is a, c, lambda, sigma
n_walkers = 50  # number of MCMC walkers
n_burn = 1000  # "burn-in" period to let chains stabilize
n_steps = 10000  # number of MCMC steps to take after burn-in
````

### Where should the walkers start?

We now have to specify the starting points of all of the walkers.  We could try a variety of starting points across parameter space.  Alternatively, if we think we have good estimates of the most probable values for the parameters, we can start the random walkers near this estimate.  We choose to do that latter.  We normally distribute our starting points around our guess with a standard deviation given by one-third of the guess value.  The exception is sigma, which we start with exponential guesses.  Note that sigma is a nuisance parameter and performing MCMC and then ignoring it is equivalent to integrating over the nuisance parameter.

````python
# Seed random numbers for reproducibility. 
np.random.seed(42)

# Generate random starting points for walkers.  
# p0[i,j] is the starting point for walk i along variable j.
# We'll randomize in a Gaussian fashion about our guesses.
p0 = np.empty((n_walkers, n_dim))
p0[:,0] = np.random.normal(data.y.max(), data.y.max() / 3.0, n_walkers) # a
p0[:,1] = np.random.normal(data.y.min(), data.y.min() / 3.0, n_walkers) # c
p0[:,2] = np.random.normal(0.2, 0.05, n_walkers)  # lambda
p0[:,3] = np.random.exponential(1.0, n_walkers)   # sigma
````

Next, we instantiate an ensemble sampler.  The syntax of `emcee` is quite straightforward.

````python
# Set up the EnsembleSampler instance
sampler = emcee.EnsembleSampler(n_walkers, n_dim, log_posterior, 
                                args=(y_model, data.x, data.y, (), jeffreys))
````

Now, we use the `run_mcmc` method of our `EmsembleSampler` instance to perform the MCMC.  The return values of `run_mcmc` are the current position of each walker, the current probability value of each walker, and the state of the random number generator.  These values are not useful, except to continue sampling.  We use them to remember where we are after the burn-in.

````python
# Do the burn-in
pos, prob, state = sampler.run_mcmc(p0, n_burn)
````

After the burn-in, we reset the sample and perform our walks.

````python
# Reset sampler and run from the burn-in state we got to
sampler.reset()
pos, prob, state = sampler.run_mcmc(pos, n_steps)
````

The results of the MCMC are stored in `sampler.chain`.  This is an `n_walkers` by `n_steps` by `n_dim` array.  It is convenient for use to reshape it into an `n_dim` by `n_walkers * n_steps` array so we can access all points sampled, say for parameter with index `0`, as `mcmc_trace[0,:]`.  This is actually required by the MCMC trace parsers we will use.

````python
# Reshape walker results for convenience
mcmc_trace = sampler.chain.reshape(-1, n_dim).transpose()
````

This is a multidimensional problem, and we applied some nice Bayesian techniques, so we should report a credible region.  People still like to see +/- intervals, though.  So, let's compute the mean values of our parameters, +/- one standard deviation.

````python
# Compute mean parameter values
p_mean = mcmc_trace.mean(axis=1)
p_std = mcmc_trace.std(axis=1)

# Show our previous results form optimization for comparison
print("""
             MCMC           JB's curve_fit_1d
        ==============       ==============
a      = {0:.3f} +- {1:.3f},     {6:.3f} +- {7:.3f}
c      = {2:.3f} +- {3:.3f},     {8:.3f} +- {9:.3f}
lambda = {4:.3f} +- {5:.3f},     {10:.3f} +- {11:.3f}
""".format(p_mean[0], p_std[0], p_mean[1], p_std[1], p_mean[2], p_std[2],
           a_star, np.log(a_err_bar), c_star, np.log(c_err_bar), 
           lam_star, lam_err_bar))
````

We see similar results.  Note, though, that the mean of the parameter values is not generally the most probable.  If a parameter are Gaussian distributed, the mean value is the most probable.  (As a side note, simply saying that a parameter has a probability distribution is a uniquely Bayesian thing to say.)  We can check from our MCMC traces.

````python
# The sampler kept track of the log of the probability at each step,
# so we get the index with the max log probability and pull out param values.
max_ind = np.argmax(sampler.lnprobability)
a_max, c_max, lam_max = mcmc_trace[:3, max_ind]

# Compute the log of the posterior at the mean
ln_post_mean = log_posterior(p_mean, y_model, data.x, data.y, (), jeffreys)

print("""
a_mean =      {0:.3f},  a_max =      {1:.3f}
c_mean =      {2:.3f},  c_max =      {3:.3f}
lambda_mean = {4:.3f},  lambda_max = {5:.3f}

lnprob at mean = {6:.3f}
lnprob at max  = {7:.3f}
""".format(mcmc_trace[0,:].mean(), a_max,
           mcmc_trace[1,:].mean(), c_max,
           mcmc_trace[2,:].mean(), lam_max,
           ln_post_mean, sampler.lnprobability.max()))
````

The means are very close to the maximal (most probable) values.  We see that the posterior probability evaluated at the mean from our traces is close to the maximal probability obtained by the walkers.  Remember, though, this is not generally the case!  We have (hopefully) sampled all of the relevant parts of the posterior probability distribution, so we do not need to make assumptions.  That took a lot of computational work, so we want to use all the information we got.

The first thing we'll do is look at all of the possible curves that might describe the data as visited by the sampler.

````python
# Pull out parameters (ignore nuisance parameter).
# We need to reshape the parameters for easy input into y_model
p = list(mcmc_trace[:3].reshape((3, mcmc_trace.shape[1], 1)))

# Compute all of the possible curves visited in the calculation
y_smooth = y_model(p, x_smooth)

# Get the mean curve
y_mean = y_smooth.mean(axis=0)

# Get the standard deviations in the positions for each point along the curve
y_std = y_smooth.std(axis=0)
````

Note that when we computed the mean curve and the standard deviations in the points along the curve, we did not just generate a curve from the mean parameter values and then curves for the mean parameter curves $\pm$ one standard deviations.  Instead, we generated all curves, and then computed the mean and standard deviations in the values of those curves at each point $x$.  These are not the same operations!  The latter is preferred because it takes into account all covariances of the parameters.  It also takes into account the differening dependencies of various parameters on data from different parts of the measured curve.

We can now plot the data with the best-fit curve (the mean of all curves we sampled), plus or minus one or two standard deviations.

````python
plt.fill_between(x_smooth, y_mean - 2.0 * y_std, y_mean + 2.0 * y_std, 
                     color='lightgray')
plt.fill_between(x_smooth, y_mean - y_std, y_mean + y_std, color='#1f78b4')
plt.plot(data.x, data.y, 'ko')
plt.plot(x_smooth, y_mean, 'r-', linewidth=1)
plt.xlabel(r'$x$')
plt.ylabel(r'$y$');
````

The red line represents the best fit curve.  The blue region is $\pm$ one standard deviations, while the gray region is $\pm$ 4 standard deviations.  We didn't bother to plot $\pm$ 2 or 3 standard deviations because, as we can see, the data strongly restrict the values the parameters may take.
