```python
# As usual, import modules
from __future__ import division, absolute_import, \
                                    print_function, unicode_literals

import numpy as np
import matplotlib.pyplot as plt

import emcee

# Necessary to display plots in this IPython notebook
%matplotlib inline 

# Use seaborn styles for nice-looking plots
import seaborn as sns
sns.set()
```

## $x$ and $y$ are Gaussian distributed

We'll make a distribution that is Gaussian distributed.

```python
def log_gauss(x, mu, sigma):
    """
    Log of unnormalized Gaussian distribution with a given mu, sigma.
    """
    return -(x - mu)**2 / 2.0 / sigma**2
```

## Generation of MCMC samples

We'll use MCMC to sample these to get chains for $x$ and $y$.  We'll take $x_0 = 10$, $\sigma_x = 2$, $y_0 = 20$, $\sigma_y = 2$.  We will also consider $y_{2,0} = 30$ and $\sigma_{y,2} = 5$ to investigate the effects of different variances.

```python
# Parametrize dists
x_0 = 10.0
sigma_x = 2.0
y_0 = 20.0
sigma_y = 2.0
y_2_0 = 30.0
sigma_y_2 = 5.0

# Set up MCMC parameters
n_dim = 1        # number of parameters in the model
n_walkers = 10   # number of MCMC walkers
n_burn = 500     # "burn-in" period to let chains stabilize
n_steps = 10000   # number of MCMC steps to take after burn-in

# Seed random number generator for reproducibility
np.random.seed(42)

# Generate random starting points for walkers.  
# p0[i,j] is the starting point for walk i along variable j.
p0 = np.empty((n_walkers, n_dim))
p0[:,0] = np.random.uniform(10.0, 20.0, n_walkers)

# Set up the EnsembleSampler instances
sampler_x = emcee.EnsembleSampler(n_walkers, n_dim, log_gauss, 
                                  args=(x_0, sigma_x))
sampler_y = emcee.EnsembleSampler(n_walkers, n_dim, log_gauss, 
                                  args=(y_0, sigma_y))
sampler_y_2 = emcee.EnsembleSampler(n_walkers, n_dim, log_gauss, 
                                  args=(y_2_0, sigma_y_2))

# Do the burn-ins and samples
pos, prob, state = sampler_x.run_mcmc(p0, n_burn, storechain=False)
sampler_x.reset()
pos, prob, state = sampler_x.run_mcmc(pos, n_steps)
    
pos, prob, state = sampler_y.run_mcmc(p0, n_burn, storechain=False)
sampler_y.reset()
pos, prob, state = sampler_y.run_mcmc(pos, n_steps)

pos, prob, state = sampler_y_2.run_mcmc(p0, n_burn, storechain=False)
sampler_y_2.reset()
pos, prob, state = sampler_y_2.run_mcmc(pos, n_steps)
```

Let's look at what we got from MCMC.

```python
# Make histogram
n_x, b, p = plt.hist(sampler_x.flatchain[:,0], histtype='step', 
                     color='k', normed=True, bins=100, lw=1)
n_y, b, p = plt.hist(sampler_y.flatchain[:,0], histtype='step', 
                     color='b', normed=True, bins=100, lw=1)
n_y_2, b, p = plt.hist(sampler_y_2.flatchain[:,0], histtype='step', 
                       color='g', normed=True, bins=100, lw=1)

# Set plot attributes
plt.xlim((0, 50))
plt.ylim((0, np.max((n_x.max(), n_y.max(), n_y_2.max()))))
plt.xlabel('$x, y$', fontsize=18)
plt.legend(('$P(x|I)$', '$P(y|I)$', '$P(y_2|I)$'), fontsize=18)
plt.tight_layout()    
```

## Computing $P(z~|~I)$

We can easily use our MCMC results to compute $P(z~|~I)$, where $z = x/y$.

```python
# Compute P(z|I)
flatchain_z = sampler_x.flatchain / sampler_y.flatchain
flatchain_z_2 = sampler_x.flatchain / sampler_y_2.flatchain

# Plot histogram (for clarity, Just keep range near peak)
inds = (flatchain_z > 0) & (flatchain_z < 1.25)
inds_2 = (flatchain_z_2 > 0) & (flatchain_z_2 < 1.0)
n, b, p = plt.hist(flatchain_z[inds], histtype='step', 
                   color='b', normed=True, bins=100, lw=1, label='$P(z|I)$')

n, b, p = plt.hist(flatchain_z_2[inds_2], histtype='step', 
                   color='g', normed=True, bins=100, lw=1, label='$P(z_2|I)$')


# Compute properties of x and y dists
x_0 = sampler_x.flatchain.mean()
sigma_x = sampler_x.flatchain.std()
y_0 = sampler_y.flatchain.mean()
sigma_y = sampler_y.flatchain.std()
y_2_0 = sampler_y_2.flatchain.mean()
sigma_y_2 = sampler_y_2.flatchain.std()


# Compute z rule of thumb
z_0 = x_0 / y_0
sigma_z = z_0 * np.sqrt(sigma_x**2 / x_0**2 + sigma_y**2 / y_0**2)
z_2_0 = x_0 / y_2_0
sigma_z_2 = z_2_0 * np.sqrt(sigma_x**2 / x_0**2 + sigma_y_2**2 / y_2_0**2)


# Make smooth Gaussian approximation
z_smooth = np.linspace(0, 1.25, 200)
z_dist = np.exp(-(z_smooth - z_0)**2 / 2.0 / sigma_z**2)\
                    / np.sqrt(2.0 * np.pi) / sigma_z
z_2_dist = np.exp(-(z_smooth - z_2_0)**2 / 2.0 / sigma_z_2**2)\
                    / np.sqrt(2.0 * np.pi) / sigma_z_2
    
# Plot the smooth distribution
plt.plot(z_smooth, z_dist, 'r-')
plt.plot(z_smooth, z_2_dist, 'r-')
plt.xlim((0, 1.25))
plt.xlabel('$z = x/y$', fontsize=18)
plt.ylabel('$P(z|I)$', fontsize=18)
plt.legend(fontsize=18)
```
